```python
class Node:
  def __init__(self,id,name,task):
    self.id = id
    self.name = name
    self.task = task
    self.next = None
    self.prev = None

class DoublyLinkedList:
  def __init__(self):
    self.head = None

    ##insertion at first
  def insert_at_first(self,id,name,task):
      new_node = Node(id,name,task)
      a=self.head
      a.prev = new_node
      new_node.next = a
      self.head = new_node

  ##insert at last
  def insert_at_last(self,id,name,task):
    new_node = Node(id,name,task)
    a = self.head
    if a is None:
      self.head = new_node
    else:
      while a.next is not None:
         a = a.next
      a.next =new_node
      new_node.prev = a
      new_node.next = None

  ## insert with index
  def insert_at_index(self,id,name,task,index):
    new_node=Node(id,name,task)
    a=self.head
    count=0

    while count<index-1:
      a=a.next
      count+=1
    new_node.next=a.next
    a.next=new_node
    new_node.prev=a
    a.next.prev=new_node

## delete first element
  def delete_first(self):
    a=self.head
    self.head=a.next
    a.next=None
    self.head.prev=None

##delete the last element

  def delete_last(self):
    a=self.head
    while a.next is not None:
      a=a.next

    a.prev.next=None
    a.prev=None

##delete  nth element
  def delete_nth_element(self,index):
    a=self.head

    count=0
    while count<index-1:
      a=a.next
      count+=1
    print("jfhgjfgj",a.name)
    a.next=a.next.next
    a.next.next.prev=a

## Search data in linked list by binary search
  def linear_search(self,name):
    a=self.head
    while a is not None:
      if a.name==name:
        return a
      a=a.next
    return None






  def fdisplay(self):
    if self.head is None:
      print("the linked list is empty")
    else:
        a = self.head

        while a is not None:
          print(a.id,"->",a.name,"->",a.task)
          a = a.next
        print("None")

  def bdisplay(self):
    if self.head is None:
      print("the linked list is empty")
    else:
      a = self.head
      while a.next is not None:
          a = a.next
      while a is not None:
        print(a.id,"->",a.name,"->",a.task)
        a = a.prev

n1 = Node(1,"A",[2,3,4,4])
n2 = Node(2,"B",[23,54,6,5])
n3 = Node(3,"C",[5,6,7,5,5])
n4 = Node(4,"D",[5,2,33,4])
dll= DoublyLinkedList()

dll.head = n1
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2
n3.next = n4
n4.prev = n3

##insert
dll.insert_at_first(0,"insert first",[2,3,4,4])
dll.insert_at_last(5,"inseeert at last",[2,3,4,4])
dll.insert_at_index(500,"insert at index",[2,3,4,4],3)

dll.fdisplay()
dll.bdisplay()
##delete
dll.delete_first()
dll.delete_last()
dll.delete_nth_element(2)

##display
print("after delete")
dll.fdisplay()
dll.bdisplay()

##search
print(dll.linear_search("C"))

```
